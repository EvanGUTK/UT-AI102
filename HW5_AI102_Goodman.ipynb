```python
print(" I could not find a a text3 or text1 so I made my own in print")
print("\npart 1")

# Define text3
text3 = ["running", "fixing", "example", "text", "think", "Python", "That", "apple"]

# Empty lists for each condition
ending_ing = []
contains_x = []
contains_th = []
titlecase_words = []
other_words = []

# Loop through words in text3
for word in text3:
    if word[0].isupper() and word[1:].islower():  
        
# Title case check 
        titlecase_words.append(word)
    elif "th" in word.lower():  
        
# Check if the word contains "th"
        contains_th.append(word)
    elif word[-3:] == "ing":  
        
# Check if the word ends with "ing"
        ending_ing.append(word)
    elif "x" in word:  
        
# Check if the word contains "x"
        contains_x.append(word)
    else:
        other_words.append(word)  
        
print("Words ending in 'ing':", ending_ing)
print("Words containing 'x':", contains_x)
print("Words containing 'th':", contains_th)
print("Words in titlecase:", titlecase_words)
print("Other words:", other_words)


print("\npart 2")

# list of words
words = ["\nHow", "much", "wood", "would", "a", "woodchuck", "chuck", "if", "a", "woodchuck", "could", "chuck", "wood"]

# Part 2 (2): Print words that begin with "wo"
print("\nWords beginning with 'wo':")
for word in words:

# Check first two letters
    if len(word) >= 2 and word[0] == "w" and word[1] == "o":
        print(word)

# Part 2 (3): Print words longer than four characters
print("\nWords longer than four characters:")
for word in words:
    if len(word) > 4:
        print(word)

print("\npart 3")

# Sample text1
text1 = ["Whale", "whale", "fish", "Whale", "shark", "whale", "WHale", "whale"]

# Convert all words to lowercase
lowercase = []
for word in text1:
    lowercase.append(word.lower())

# Counts of "whale"
whale_count = 0
for word in lowercase:
    if word == "whale":
        whale_count += 1

# Calculate percentage
total_words = len(text1)
# Convert to percentage
percentage = (whale_count / total_words) * 100

# Print result using f string
print(f"\n Word'whale' occurs {whale_count} times, making up {percentage:.2f}% of the text.")
```
