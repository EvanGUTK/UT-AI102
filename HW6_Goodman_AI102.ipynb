```python
#Importing proper library
import nltk
from nltk.book import text2
from collections import Counter

#All conversions
words = [word.lower() for word in text2 if word.isalpha()]

word_counts = Counter(words)

unique = [word for word, count in word_counts.items() if count >= 10]

print("Unique words occurring at least 10 times:")
print(unique)

top_20 = word_counts.most_common(20)
print("\nTop 20 most frequent words and their counts:")
for word, count in top_20:
    print(f"{word}: {count}")

long = {word: count for word, count in word_counts.items() if len(word) >= 10}
print("\nWords that are at least 10 characters long and their counts:")
for word, count in long.items():
    print(f"{word}: {count}")

filtered = [word for word, count in long.items() if count >= 2]

print("\nWords that are 10+ character long, occur at least twice, sorted from most frequent to least:")
for word, count in filtered:
    print(f"{word}: {count}")
```
