```python
from textblob import TextBlob
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from flair.models import TextClassifier
from flair.data import Sentence

# List of review for Aquaphor
reviews = [
    "Great product. Easily appllies, softens well, skin feels nourished.",
    "Very smooth product i love it Advise to use it under eye before concealer giving amazing results thanks",
    "The one I bought at the store, lasted me several months. The one on Amazon, lasted two weeks. I ordered another one just to see. I was right LOL. I’ll take my lazy ass to the store. SMH Amazon.",
    "Whoever thought to put this must have product in this handy stick form is a genius. I use it on my face and anywhere needed. No more gooey jars needed for simple everyday uses!",
    "I got this on February 6th and used it on my hands once or twice a day on my hands. Today I'm down to the plastic.I like this application but it's not worth it."
]

# Results
print("Results:")
for review in reviews:
    blob = TextBlob(review)
    print(review)
    print("Polarity:", blob.sentiment.polarity)
    print("Subjectivity:", blob.sentiment.subjectivity)
    print()

# NLTK
print("NLTK VADER RESULTS:")
vader = SentimentIntensityAnalyzer()
for review in reviews:
    score = vader.polarity_scores(review)
    print(review)
    print(score)
    print()

# Flair
print("FLAIR RESULTS:")
classifier = TextClassifier.load('sentiment')
for review in reviews:
    sentence = Sentence(review)
    classifier.predict(sentence)
    print(review)
    print(sentence.labels)
    print()
```
